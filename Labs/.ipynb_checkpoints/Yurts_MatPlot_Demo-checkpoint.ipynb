```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook

fir_fig = plt.figure(1)
week = pd.Series({"Monday":10, "Tuesday":31, "Wednesday":53, "Thursday":47, "Friday":66})
plt.title("Моя тривожність протягом тижня")
plt.plot(week, color="r")
fir_fig.show()
```

```python
sec_fig = plt.figure(2)
week = pd.Series({"Samsung":35, "Xiaomi":26, "Apple":29, "Huawei":10})
plt.title("Об'єм продаж компаній на ринку мобільних телефонів")
plt.pie(week, labels=week.index)
sec_fig.show()
```

```python
import random as rnd

thir_fig = plt.figure(3)

plt.plot([rnd.randint(0,10) for i in range(0, 10)], marker="o", label="some data 1")
plt.plot([rnd.randint(0,10) for i in range(0, 10)], marker="*", label="some data 2")

plt.legend()
plt.title("Співставлення двох списків випадкових значень")
thir_fig.show()
```

```python
forth_fig = plt.figure(4)

plt.scatter(list(np.random.randint(1, 11, size = 30)), list(np.random.randint(1, 11, size = 30)), s=55, marker="+")
plt.scatter(list(np.random.randint(1, 11, size = 30)), list(np.random.randint(1, 11, size = 30)))
plt.title("Випадкові точки")

forth_fig.show()
```

```python
fif_fig = plt.figure(5)
x = np.arange(1, 16)
y1 = np.random.randint(1, 10, size = 15)
y2 = np.random.randint(1, 10, size = 15)
plt.bar(x - 0.2, y1, width = 0.4, color="#10ac84")
plt.bar(x + 0.2, y2, width = 0.4, color="#ff6b6b")

fif_fig.show()
```
