```python
import pandas as pd
```

```python
df = pd.read_csv('olympics.csv', index_col = 0, skiprows=1)
df.head()
```

```python
df.rename(columns={"01 !": "Gold", "02 !": "Silver", "03 !": "Bronze"}, inplace=True)
df.head()
```

```python
df[['Gold', 'Silver', 'Bronze']]
```

```python
df = df.drop('Totals')
```

```python
df[df['Gold']>0]
```

```python
df[df['Gold']>0].shape[0]
```

```python
df[df['Gold'] == df['Gold'].max()]
```

```python
df[ (df['Gold'] > 0) & (df['Silver'] > 0) ]
```

```python
df['country'] = df.index
df.index = df['Gold']
df
```

```python
df[df['country'] == 'Ukraine\xa0(UKR)']
```

```python
df['sum_medals'] = df.apply(lambda row: row['Gold'] + row['Silver'] + row['Bronze'], axis=1)
df
```

```python
df = df.sort_values(by='sum_medals', ascending=False)
df
```

```python
import matplotlib.pyplot as plt

plt.bar(df['country'][:10], df['sum_medals'][:10])
plt.show()
```
