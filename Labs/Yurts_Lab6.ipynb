```python
import math

#================== ЗАВДАННЯ 1 ==================

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return '(' + str(self.x) + str(self.x) + ', ' + str(self.y) + ')'

    def __eq__(self, other):
        return (other.x == self.x) and (other.y == self.y)

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Point(self.x - other.x, self.y - other.y)


class Vector:
    def __init__(self, point_first, point_second):
        self.p_fir = point_first
        self.p_sec = point_second

    def __str__(self):
        return "First point: " + str(self.p_fir) + ", Second point: " + str(self.p_sec)

    def __eq__(self, other):
        return (other.p_fir == self.p_fir) and (other.p_sec == self.p_sec)

    def __add__(self, other):
        return Vector(self.p_fir + other.p_fir, self.p_sec + other.p_sec)

    def leng(self):
        diff = self.p_fir - self.p_sec
        return math.sqrt(pow(diff.x, 2) + pow(diff.y, 2))

    def scal(self, other):
        return self + other

    def perpend(self, other):
        return self.scal(other) == Vector(Point(0, 0), Point(0, 0))


vec_first = Vector(Point(5, 3), Point(2, 6))
vec_second = Vector(Point(3, 5), Point(6, 9))

print('Length: \t' + str(vec_first.leng()))
print('Scalar: \t' + str(vec_first.scal(vec_second)))
print('Are perpendicular: \t' + str(vec_first.perpend(vec_second)))
```

```python
#================== ЗАВДАННЯ 2 ===================

class Student:
    def __init__(self, surname, name, marks):
        self.surname = surname
        self.name = name
        self.marks = marks

    def __str__(self):
        return self.name + " " + self.surname + " | Subjects: " + str(list(self.marks))

    def subjects(self):
        return ', '.join(list(self.marks))

    def marks_for_subject(self, subject):
        return self.marks[subject]

    def average_mark_for_subject(self, subject):
        return sum(self.marks[subject]) / len(self.marks[subject])

    def average_mark(self):
        return sum([sum(marks[subj])/len(marks[subj]) for subj in self.marks.keys()])/len(marks)

    def passed_subject(self, subject):
        return self.average_mark_for_subject(subject) >= 4

    def add_subject(self, subject, marks):
        self.marks[subject] = marks

    def add_mark_to_subject(self, subject, mark):
        self.marks[subject].append(mark)

    def print_info(self):
        print('--------All info--------\n' + str(self))
        for subj in self.marks.keys():
            print("Average mark for "+ subj + " is " + str(self.average_mark_for_subject(subj)))
        print('------------------------')  



marks = {'math':[5,4,4,4], 'data_bases':[4,4,5,3,5], 'economy':[2,3,2,4,5,4]}

stud = Student('Yurts', 'Taras', marks)

print('Subjects: ' + str(stud.subjects()))
print(''str(stud.marks_for_subject('data_bases')))
print('Avarage mark for math: ' + str(stud.average_mark_for_subject('math')))
print('General avarage mark: ' + str(stud.average_mark()))
print('Passed economy: ' + str(stud.passed_subject('economy')))

new_subject_name = 'microprocessing'
new_subject_marks = [4,2,5,5,4,5]
stud.add_subject(new_subject_name, new_subject_marks)

stud.add_mark_to_subject('math', 4)

stud.print_info()
```
