```python
# ====  Основне завдання  ====

# Лісти

from abc import ABC, abstractmethod 

# абстрактний клас, що містить базову функціональність для моїх імплементацій доступу до ліста чи словника

class MyCollection(ABC):
    coll = None
    available_commands = ''
    
    #     австрактний метод
    def init_collection(self):
        pass

    #     австрактний метод
    def show_available_commands(self):
        print(self.available_commands)
    
    #     австрактний метод
    def add_element(self):
        pass
    
    #     австрактний метод
    def remove_element(self):
        pass
    
    def show(self):
        print(self.coll)

        
        
# класс, що наслідується від вище вказаного
        
class MyList(MyCollection):
    
    def init_collection(self):
        self.coll = list()
        print("Done")
        
    def create_natural_range(self):
        try:
            self.coll = list(range(1,int(input('Довжина списку натуральних чисел: '))+1))
        except ValueError:
            print('Неправильний тип значення!')
        
    def add_element(self):
        elemenet_to_add = input('Element: ')
        try:
            self.coll.append(elemenet_to_add)
        except:
            print("Список ще не створений!")
    
    def remove_element(self):
        try:
            index_to_delete = int(input('Індекс елемента: '))
            if(0 <= index_to_delete < len(self.coll)):
                self.coll.pop(index_to_delete)
            else:
                print('Дане число не входить в межі масиву!')
        except ValueError:
            print('Неправильний тип значення!')

def exit():
    return -1

collection = MyList()
collection.available_commands = 'My menu\n1: create list\n2: create natural range\n3: add element\n4: preview\n5: remove element\n6: exit'
collection.show_available_commands()

switcher = {
    1 : collection.init_collection,
    2 : collection.create_natural_range,
    3 : collection.add_element,
    4 : collection.show,
    5 : collection.remove_element,
    6 : exit
}

while(True):
    try:
        operation_num = int(input('=======\nВведіть бажану команду: '))
        if 0 < operation_num <= len(switcher):
            if switcher[operation_num]() == -1:
                break
        else:
            print('Дане число не входить в пункти меню!')
    except ValueError:
        print('Неправильний тип значення!')            
        
```

```python

#Словники

class MyDict(MyCollection):
    
    def init_collection(self):
        self.coll = dict()
        print("Done")
        
    def add_element(self):
        new_item_key = input('Ключ нового елемента: ')
        try:
            if new_item_key in self.coll and input('Значення за таким ключем вже присутнє. Перезаписати? Y/N: ').upper() == 'N':
                return
            self.coll[new_item_key] = input('Значення нового елемента: ')
        except:
            print('Словник ще не створений')
    
    def remove_element(self):
        item_key = input('Ключ елемента: ')
        try:
            if item_key in self.coll:
                del self.coll[item_key]
            else:
                print('Значення за таким ключем не існує!')
        except:
            print('Словник ще не створений')

collection = MyDict()
collection.available_commands = 'My menu\n1: create dictionary\n2: add element\n3: preview\n4: remove element\n5: exit'
collection.show_available_commands()

switcher = {
    1 : collection.init_collection,
    2 : collection.add_element,
    3 : collection.show,
    4 : collection.remove_element,
    5 : exit
}

while(True):
    try:
        operation_num = int(input('=======\nВведіть бажану команду: '))
        if 0 < operation_num <= len(switcher):
            if switcher[operation_num]() == -1:
                break
        else:
            print('Дане число не входить в пункти меню!')
    except ValueError:
        print('Неправильний тип значення!')            
```
